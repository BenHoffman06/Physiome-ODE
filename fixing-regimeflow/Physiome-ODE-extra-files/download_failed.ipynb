```python
from pathlib import Path
from urllib.parse import urlsplit, urlunsplit
from urllib.request import urlretrieve
```

```python
def strip_to_cellml(url: str) -> str:
    """
    Keep only the part of the URL up to and including '.cellml'.
    """
    url = url.strip()
    parts = urlsplit(url)

    path = parts.path
    idx = path.find(".cellml")
    if idx == -1:
        raise ValueError(f"No .cellml found in URL: {url}")

    stripped_path = path[:idx + len(".cellml")]
    return urlunsplit((parts.scheme, parts.netloc, stripped_path, "", ""))
```

```python
strip_to_cellml("https://models.physiomeproject.org/exposure/6c66cc4b25a834b256b9dbb23a3dd9e4/albrecht_colegrove_friel_2002.cellml/@@cellml_codegen/Python")
```

```python
def download_cellmls_from_file(link_file: str, out_dir: str = "cellml_files"):
    out_path = Path(out_dir)
    out_path.mkdir(parents=True, exist_ok=True)

    with open(link_file, "r") as f:
        urls = [line.strip() for line in f if line.strip()]

    for url in urls:
        try:
            cellml_url = strip_to_cellml(url)
            filename = Path(urlsplit(cellml_url).path).name
            save_path = out_path / filename

            print(f"Downloading: {cellml_url}")
            urlretrieve(cellml_url, save_path)
            print(f"Saved to: {save_path}")
        except Exception as e:
            print(f"Failed for {url}\n  {e}")
```
