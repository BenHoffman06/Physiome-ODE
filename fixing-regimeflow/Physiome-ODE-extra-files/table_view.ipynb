```python
import pandas as pd
```

```python
df = pd.read_csv("CellML_model_list_updates.csv")
```

```python
# Assumes your CSV is already loaded as df
# If not, uncomment: import pandas as pd

df.columns = df.columns.str.strip()

has_pde_and_ode = ((df["PDE count"] > 0) & (df["ODE count"] > 0)).sum()
has_ode = (df["ODE count"] > 0).sum()
has_algebraic = (df["Algebraic Equation Count"] > 0).sum()
only_algebraic = (
    (df["Algebraic Equation Count"] > 0) &
    (df["ODE count"] == 0) &
    (df["PDE count"] == 0)
).sum()

print(f"Has PDE and ODE: {has_pde_and_ode}")
print(f"Has ODE: {has_ode}")
print(f"Has Algebraic: {has_algebraic}")
print(f"Only Algebraic: {only_algebraic}")
```
